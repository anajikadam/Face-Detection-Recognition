```python
def load_image( path ):
    img_base = cv2.imread( path )
    print(img_base.shape)
    img_proc = cv2.cvtColor(img_base, cv2.COLOR_BGR2RGB)
    return img_proc
```

```python
path = r"/content/test23.jpg"
img = load_image( path )

center = (246, 412)
cv2.circle(img, center, 2, (255, 255, 255), thickness = 2)
cv2.line(img, (196,453), (296,462), (255, 255, 255), thickness = 2)

fig = plt.figure(figsize = (15, 15))
plt.axis('off')
plt.imshow(img)
plt.show()
```
